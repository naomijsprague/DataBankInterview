```python
import pandas as pd
import numpy
```

```python
mock_file = pd.read_csv('sales by member detail 2013-2015 QB.csv')
```

```python
mock_file = mock_file[mock_file.Num > 0]
mock_file
```

```python
denote_company = set(['Network', 'LLC', 'LLP', 'Minnesota', 'Inc', 'INC', 'Bank', 'Engineers', 'Companies', 'Corporation','\'s', 'Quidditch', 'Plastics'])
```

```python
mock_file['Title'] = ''
mock_file['First Name'] = ''
mock_file['Title'] = ''
mock_file['Middle Initial'] = ''
mock_file['Last Name'] = ''
mock_file['Suffix'] = ''
mock_file['Company'] = ''
mock_file['Salutation'] = ''
mock_file['Category'] = ''
mock_file['Sub-Category'] = ''
mock_file['Item'] = ''



mock_file
```

```python
mock_file = mock_file.to_dict()
```

```python
def is_person_test(key):     # this function is to separate keys as 'raw' strings into company or name values
    if any(st in key for st in denote_company):
        return False
    else:
        return True

titles = set(['Dr', 'Dr.', 'Mr', 'Mr.', 'Ms', 'Ms.', 'Mrs', 'Mrs.', 'Sir', 'Dame'])
def title_test(x): # x is a list of strings
    if any(st in x for st in titles):
        return True
    else:
        return False


suffixes = set(['MD', 'PhD', 'Ph.D', 'Jr', 'Jr.'])
def suffix_test(x): #x is a list of strings
    if any(st in x for st in suffixes):
        return True
    else:
        return False

def two_humans_test(x): # x is a list of strings
    two_humans = ['&','and']
    if any(st in x for st in two_humans):
        return True
    else:
        return False
    
def surname_first_test(x): # x is a list of strings
    if ',' in x[0]:
        return True
    else: 
        return False


for value, key in mock_file['Name'].iteritems():
    if is_person_test(key) == True:
        company = None
        current_name = key.strip() #remove leading and trailing space/newline characters
        current_name = current_name.split(' ')  #split string into list of strings on spaces
        if title_test(current_name) == True:
            for i in current_name:
                if i in titles:
                    line_title = i               #save line_title for final output
            current_name = [x for x in current_name if x not in titles]
        else:
            line_title = None
        if suffix_test(current_name) == True:
            for i in current_name:
                if i in suffixes:
                    line_suffix = i               #save line_suffix for final output
            current_name = [x for x in current_name if x not in suffixes]
        else:
            line_suffix = None
        if surname_first_test(current_name) == True:
            lastname = current_name[0].strip(',') #save lastname for final output
            if two_humans_test(current_name) == True:
                firstname = current_name[1]+ ' '+current_name[2]+' '+current_name[3]
                if len(current_name) > 4:
                    middle_initial = current_name[4][0]
                else:
                    middle_initial = None
            else:
                firstname = current_name[1]
                if len(current_name)>2:
                    middle_initial = current_name[2]
                else: middle_initial = None
        if surname_first_test(current_name) == False:
            if len(current_name) == 3:
                firstname = current_name[0]
                middle_initial = current_name[1][0]
                lastname = current_name[2]
            if len(current_name) == 2:
                firstname = current_name[0]
                lastname = current_name[1]
                middle_initial = None
        salutation = 'Dear' 
        if line_title != None:
            salutation = salutation +' '+ line_title
        if firstname != None:
            salutation = salutation +' '+ firstname
        if middle_initial != None:
            salutation = salutation +' '+ middle_initial
        if lastname != None:
            salutation = salutation +' '+ lastname
        if line_suffix != None:
            salutation = salutation +' '+ line_suffix
            
    else: 
        line_title = None
        firstname = None
        middle_initial = None
        lastname = None
        line_suffix = None
        company = key
        salutation = 'Dear' + ' ' + company
    mock_file['Title'][value] = line_title
    mock_file['First Name'][value] = firstname
    mock_file['Middle Initial'][value] = middle_initial
    mock_file['Last Name'][value] = lastname
    mock_file['Suffix'][value] = line_suffix
    mock_file['Company'][value] = company
    mock_file['Salutation'][value] = salutation
    
for key, value in mock_file['Product/Service'].iteritems():
    contents = value.split(':')
    if len(contents) == 1:
        category = None
        sub_category = None
        item = contents
    if len(contents) == 2:
        category = contents[0]
        sub_category = None
        item = contents[1]
    if len(contents) == 3:
        category = contents[0]
        sub_category = contents[1]
        item = contents [2]
    mock_file['Category'][key] = category
    mock_file['Sub-Category'][key] = sub_category
    mock_file['Item'][key] = item
    
```

```python
data_frame = pd.DataFrame(mock_file)
```

```python
data_frame["Member ID"] = data_frame["Name"].astype("category") #Categories are new to me but so useful
data_frame['Member ID'] #this shows me there are 140 categories, which means 140 Members that need IDs
```

```python
data_frame["Member ID"].cat.categories = range(1,141)  #I'm setting the categories to have a number from 1 to 141
data_frame = data_frame.set_index(data_frame['Member ID'])
```

```python
Members_df = data_frame[['Title', 'First Name', 'Middle Initial', 'Last Name', 'Suffix', 'Company', 'Salutation']]
Members_df = Members_df.drop_duplicates()
```

```python
Members_df.to_csv('Members.csv') #Creates CSV File 1
```
