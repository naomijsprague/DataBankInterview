```python
mock_file = pd.read_csv('sales by member detail 2013-2015 QB.csv')
```

```python
mock_file = mock_file[mock_file.Num > 0]
mock_file
```

```python
list_of_names = []
for i in mock_file.index:
    list_of_names.append(mock_file.ix[i]['Name'])
```

```python
denote_company = set(['Network', 'LLC', 'LLP', 'Minnesota', 'Inc', 'INC', 'Bank', 'Engineers', 'Companies', 'Corporation','\'s', 'Quidditch', 'Plastics'])
```

```python
companies = set()
for i in list_of_names:
    if any(st in i for st in denote_company):
        companies.add(i)
        
is_person = []
for i in mock_file.index:
    if mock_file.ix[i]['Name'] in companies:
        is_person.append(False)
    else:
        is_person.append(True)
        
mock_file['person'] = is_person
```

```python
mock_file['Title'] = ''
mock_file['First Name'] = ''
mock_file['Title'] = ''
mock_file['Middle Initial'] = ''
mock_file['Last Name'] = ''
mock_file['Suffix'] = ''
mock_file['Company'] = ''
mock_file['Salutation'] = ''


mock_file.Name.apply(str)
mock_file
```

```python
titles = set(['Dr', 'Dr.', 'Mr', 'Mr.', 'Ms', 'Ms.', 'Mrs', 'Mrs.', 'Sir', 'Dame'])
def title_test(x): # x is a list of strings
    if any(st in x for st in titles):
        return True
    else:
        return False


suffixes = set(['MD', 'PhD', 'Ph.D', 'Jr', 'Jr.'])
def suffix_test(x): #x is a list of strings
    if any(st in x for st in suffixes):
        return True
    else:
        return False

def two_humans_test(x): # x is a list of strings
    two_humans = ['&','and']
    if any(st in x for st in two_humans):
        return True
    else:
        return False
    
def surname_first_test(x): # x is a list of strings
    if ',' in x[0]:
        return True
    else: 
        return False

members_dict = {}

for i, row in mock_file.iterrows():
    if mock_file.iloc[i]['person'] == True:
        company = None
        current_name = mock_file.ix[i]['Name'].strip() #remove leading and trailing space/newline characters
        current_name = current_name.split(' ')  #split string into list of strings on spaces
        if title_test(current_name) == True:
            for i in current_name:
                if i in titles:
                    line_title = i               #save line_title for final output
            current_name = [x for x in current_name if x not in titles]
        else:
            line_title = None
        if suffix_test(current_name) == True:
            for i in current_name:
                if i in suffixes:
                    line_suffix = i               #save line_suffix for final output
            current_name = [x for x in current_name if x not in suffixes]
        else:
            line_suffix = None
        if surname_first_test(current_name) == True:
            lastname = current_name[0].strip(',') #save lastname for final output
            if two_humans_test(current_name) == True:
                firstname = current_name[1]+ ' '+current_name[2]+' '+current_name[3]
                if len(current_name) > 4:
                    middle_initial = current_name[4][0]
                else:
                    middle_initial = None
        if surname_first_test(current_name) == False:
            if len(current_name) == 3:
                firstname = current_name[0]
                middle_initial = current_name[1][0]
                lastname = current_name[2]
            if len(current_name) == 2:
                firstname = current_name[0]
                lastname = current_name[1]
                middle_initial = None
        salutation = 'Dear' 
        if line_title != None:
            salutation = salutation +' '+ line_title
        if firstname != None:
            salutation = salutation +' '+ firstname
        if middle_initial != None:
            salutation = salutation +' '+ middle_initial
        if lastname != None:
            salutation = salutation +' '+ lastname
        if line_suffix != None:
            salutation = salutation +' '+ line_suffix
            
    else: 
        line_title = None
        firstname = None
        middle_initial = None
        lastname = None
        line_suffix = None
        company = str(mock_file.iloc[i]['Name'])
        salutation = 'Dear' + ' ' + company
    mock_file.set_value(i,'Title', line_title)
    mock_file.set_value(i, 'First Name', firstname)
    mock_file.set_value(i, 'Middle Initial', middle_initial)
    mock_file.set_value(i, 'Last Name', lastname)
    mock_file.set_value(i, 'Suffix', line_suffix)
    mock_file.set_value(i, 'Company', company)
    mock_file.set_value(i, 'Salutation', salutation)
    
        
```
